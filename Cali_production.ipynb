```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
file = 'Resources/2020CaliforniaOilAndGasWells.csv'
file_2='Resources/2020Production_monthly.csv'
file_3='Resources/2019Production_monthly.csv'
file_4='Resources/2018Production_monthly.csv'

```

```python
# The correct encoding must be used to read the CSV in pandas
wells = pd.read_csv(file)
df_20=pd.read_csv(file_2)
df_19=pd.read_csv(file_3)
df_18=pd.read_csv(file_4)
df_20
wells
```

```python
# 2018 production
df_18_sort=df_18[['APINumber', 'FieldCode', 'ProductionReportDate', 'APIGravityofOil', 'OilorCondensateProduced']]
df_18_sort['OilorCondensateProduced'].sum()
```

```python
# 2019 production
df_19_sort=df_19[['APINumber', 'FieldCode', 'ProductionReportDate', 'APIGravityofOil', 'OilorCondensateProduced']]
df_19_sort['OilorCondensateProduced'].sum()
```

```python
# 2020 production
df_20_sort=df_20[['APINumber', 'FieldCode', 'ProductionReportDate', 'APIGravityofOil', 'OilorCondensateProduced']]
df_20_sort
```

```python
# combine production tables 2018-2020
total=df_20_sort.append([df_19_sort, df_18_sort])
total['OilorCondensateProduced'].sum()
```

```python
# refine well source data
df_rename=wells.rename(columns={'API': 'APINumber'})
df_source=df_rename.drop(['AreaCode', 'PoolCode', 'PoolName', 'AreaName', 'WellNumber', 'Section', 'SubSection', 'Township',
                'Range', 'BM', 'Operatorcode', 'OperatorStatus', 'SystemEntryDate', 'PoolWellTypeStatus'], axis=1)
df_source.head()
```

```python
# drop duplicate well APINumbers (usually associated with different pools of drilling -- not relevant to this analysis)
final_wells=df_source.drop_duplicates(['APINumber'])
final_wells.head()
```

```python
# merge final_well data with annual production by APINumber

df_merge=pd.merge(total, final_wells, how='left', on=['APINumber', 'FieldCode'])
df_merge['LeaseName'].fillna('NotFound', inplace=True)
df_merge['APIGravityofOil'].fillna(0, inplace=True)
df_merge['OilorCondensateProduced'].fillna(0, inplace=True)

df_merge.head()
```

```python
production=df_merge[['ProductionReportDate', 'OilorCondensateProduced', 'District', 'County', 'FieldName', 'LeaseName', 'OperatorName']]
production
```

```python
production_may20=production.groupby(['FieldName', 'OperatorName', 'LeaseName', 'District', 'County', 'ProductionReportDate']).sum()
production_may20.to_csv('Cali_production20.csv')
```
